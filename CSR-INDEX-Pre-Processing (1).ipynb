### If we look at the Decsription of population Density- The mean is affected by 3 outlier values and is thus higher than its median value.
### We replace the NaN Values of Population Density by its median but in Densely populated Countries like China , I will manually Change the value to a much higher mean value.

```python
#Impute the nan values in Population Density
median=SOC.PopulationDensity.median()

def impute_nan(df,variable,median):
    df[variable+"_median"]=df[variable].fillna(median)

impute_nan(SOC,'PopulationDensity',median)
SOC.head()
#Please Scroll towards the end where we have created a new column with Median values
```

### Repeat the Process with Undernourishment where NaN values will be replaced by the median as the mean value is affected by the outliers.

### In Readiness Column we replace the NaN with the Mean Values

```python
#Impute the nan values in Undernourishment
median=SOC.Undernourishment.median()

def impute_nan(df,variable,median):
    df[variable+"_median"]=df[variable].fillna(median)

impute_nan(SOC,'Undernourishment',median)
SOC.head()
#Please Scroll towards the end where we have created a new column with Median values
```

```python
#Impute the NaN values in Readiness

mean=SOC.Readiness.mean()

def impute_nan(df,variable,mean):
    df[variable+"_mean"]=df[variable].fillna(mean)

impute_nan(SOC,'Readiness',mean)
SOC.head()
#Please Scroll towards the end where we have created a new column with Median values
```

# z  =  (x-x_mean)/std
